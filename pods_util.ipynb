```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cpu_mem_usage(df):

    df['cpu_request'] = df['cpu_request'].str.replace('m', '').astype(int)
    df['cpu_usage'] = df['cpu_usage'].str.replace('m', '').astype(int)

    # Convert 'memory_request' and 'memory_usage' from string to integer values (MiB)
    df['memory_request'] = df['memory_request'].str.replace('Mi', '').astype(int)
    df['memory_usage'] = df['memory_usage'].str.replace('Mi', '').astype(int)

    # Calculate CPU usage percentage
    df['cpu_usage_percentage'] = (df['cpu_usage'] / df['cpu_request']) * 100

    # Calculate memory usage percentage
    df['memory_usage_percentage'] = (df['memory_usage'] / df['memory_request']) * 100

    return df
```

# Motivation

```python
df = [dfn0, dfn1, dfn15, dfs0, dfs1, dfs15]
df = pd.concat(df, axis=0)
df.reset_index(drop=True, inplace=True)
df
```

```python
plt.figure(figsize=(12, 6))

# Separate data for each model
models = df['model'].unique()
bar_width = 0.2  # Width of the bars
positions = []

for i, model in enumerate(models):
    model_data = df[df['model'] == model]
    x = model_data['cpu_request'] + i * bar_width  # Offset each model's bars
    positions.append(x)
    plt.bar(x, model_data['cpu_usage_percentage'], width=bar_width, label=model)

# Adding labels and title
plt.xlabel('Container Core Request (millicores)')
plt.ylabel('CPU Usage Percentage')
plt.title('CPU Usage Percentage vs. Container Core Request')
plt.xticks([r + bar_width for r in range(len(df['cpu_request'].unique()))], df['cpu_request'].unique())
plt.legend()
plt.grid(True)
plt.show()
```
